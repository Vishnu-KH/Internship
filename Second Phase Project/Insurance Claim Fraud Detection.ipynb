```python
label_encoder = LabelEncoder()
icf['fraud_reported'] = label_encoder.fit_transform(icf['fraud_reported'])
```

Converting the above column into binary columns.

```python
icf.drop(['policy_bind_date', 'insured_zip', 'incident_date', 'incident_location'], axis=1, inplace=True)
```

Dropping the above columns from the dataframe.

```python
X = icf.drop('fraud_reported', axis=1)
y = icf['fraud_reported']
```

```python
X_encoded = pd.get_dummies(X)
```

Splitting the dataset for training and testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)
y_pred_rf_classifier = rf_classifier.predict(X_test)
accuracy_rf_classifier = accuracy_score(y_test, y_pred_rf_classifier)
print("Random Forest Classifier Accuracy:", accuracy_rf_classifier)
```
